# Банки — Исследование оттока клиентов

## Описание проекта

Проведен анализ данных о клиентах регионального банка "Метанпром" с офисами в Ярославле, Ростове Великом и Рыбинске. Проведена сегментация клиентов в зависимости от числа потребляемых продуктов банка. Для каждого сегмента выданы рекомендации по продуктам, которые потенциально могут снизить отток клиентов данного сегмента.   

По итогам исследования проведена сегментация в зависимости от количества используемых продуктов, дана характеристика отточных клиентов каждого сегмента и подготовлены рекомендации по снижению оттока.

Для анализа был использован датасет, содержащий следующие данные:

- `userid` — идентификатор пользователя,
- `score` — баллы кредитного скоринга,
- `City` — город,
- `Gender` — пол,
- `Age` — возраст,
- `Equity` — количество объектов в собственности,
- `Balance` — баланс на счёте,
- `Products` — количество продуктов, которыми пользуется клиент,
- `CreditCard` — есть ли кредитная карта,
- `LastActivity` — активный клиент,
- `Estimated_salary` — заработная плата клиента,
- `Churn` — 1 - ушёл или 0 - нет.

Столбец с информацией о балансе имеет значительное количество пропусков. Логично предположить, что пропуск означает отсутствие баланса. Проверим, есть ли записи, у которых баланс напрямую указан нулевым:

```python
df[df['balance'] == 0]
```

Да есть, но их всего две. Нулевой баланс на счете может образоваться разными способами - деньги можно все потратить, а можно открыть счет, не внести средства, а затем никогда им не воспользоваться. И система может проставить условный "ноль" на балансе разными способами.    
Посмотрим на данные с пропусками подробнее чтобы попытаться предположить их происхождение.

```python
"""Зададим функцию, которая сгруппирует данные по признаку и посчитает долю пропусков в данных о балансе"""

def missed_balance (feature):
    temp = df.groupby(feature).agg({'userid':'count','balance':'count'}).reset_index()
    temp['%_missing'] = round((1 - temp['balance']/temp['userid'])*100,2)
    return temp
```

Посмотрим на распределение процента пропусков данных о балансе для категориальных переменных на столбчатом графике, а для непрерывных на диаграмме рассеяния:

```python
categorial = ['city', 'gender', 'equity', 'products', 'credit_card', 'last_activity', 'churn']
fig, ax = plt.subplots(nrows = 1, ncols = 7, figsize=(25,5))
for i, item in enumerate(categorial):
    ax[i].bar(item, '%_missing', data=missed_balance(item))
    ax[i].set_title(item)
fig.suptitle('Зависимость процента пропусков в балансе от других параметров', fontsize=15);
```

```python
continous = ['score', 'age','est_salary']
fig, ax = plt.subplots(nrows = 1, ncols = 3, figsize=(25,5))
for i, item in enumerate(continous):
    ax[i].scatter(item, '%_missing', data=missed_balance(item))
    ax[i].set_title(item)
fig.suptitle('Зависимость процента пропусков в балансе от других параметров', fontsize=15);
```

- По городам - чем больше город, тем больше процент пропусков, но нет такого, что пропуски идут из одного места. 
- По мужчинам и женщинам - ничего особо интересного. 
- По объектам в собственности - только у тех у кого ничего нет или 1 объект. 
- По продуктам ноль - выброс, в основном пропуски у клиентов с одним продуктом. 
- У четверти владельцев кредитки. 
- Почти совсем нет пропусков у ушедших клиентов.

Любопытно распределение пропусков в зависимости от кредитного рейтинга - 100% пропусков для низкого кредитного рейтинга, ноль в диапазоне 750-870 и снова всплеск.   
По возрасту и зарплате никаких закономерностей не прослеживается.   

Можно предположить, что NaN означает отрицательный баланс:
- 80% NaN у клиентов без собственности - они живут в кредит
- 60% пропусков у пользователя всего одного продукта - кредитной карты
- пропуски у клиентов не имеющих кредитки могут означать задолженность по другим видам кредита
- кредит или кредитная карта держит клиента в банке - среди ушедших доля пропусков близка к нулю
- клиенты с низким рейтингом и получили этот рейтинг из-за просрочек по кредиту. А те кто его регулярно в срок гасят, наоборот имеют высокий рейтинг в районе 950

В любом случае, верно наше предположение об отрицательном балансе или нет, предполагать, что пропуски кроют за собой какую-то положительную цифру, оснований нет. Делаем допущение, и отметим его в выводах, что пропуск данных в столбце `balance` означает нулевой баланс и соответственно заполним эти пропуски нулями.

```python
df['balance'] = df['balance'].fillna(0)
```

### Исследовательский анализ данных

- Исследование распределения признаков по датасету и их основных статистик
- Исследование статистик признаков в разрезе оттока клиентов (по признаку Churn)
- Построение матрицы корреляций признаков

Отдельно рассмотрим распределение каждого признака, оценим качество данных, возможные аномалии. Так как конечная цель нашего исследования - всказать рекомендации по уменьшению оттока клиентов, каждый признак исследуем также в разрезе параметре `churn` - ушедший/неушедший клиент.

Посмотрим на распределение непрерывных признаков на диаграмме размаха, а для категориальных - на столбчатой диаграмме. Отдельно для каждого признака построим его распределение в разрезе ушедших и оставшихся клиентов

```python
''' Для категориальных переменных мы будем проводить группировку датафрейма по переменной,
затем считать долю ушедших и смотреть на распределение по переменной с сортировкой по доле ушедших
Для этого напишем функцию, которая принимает на вход датафрейм и поле с категориальной переменной, 
группирует датафрейм по нему,считает количество ушедших клиентов и добавляет столбец с долей ушедших клиентов.
Возвращает сгруппированный датасет, отсортированный по доле ушедших клиентов'''

def grouper (data, field):
    grouped = (data.groupby(field)
    .agg({'userid':'count', 'churn':'sum'})
    .reset_index()
    .sort_values(by='userid', ascending=False))
    grouped['%_churn'] = grouped['churn']/grouped['userid']
    grouped = grouped.sort_values('%_churn')
    return grouped
```

Нам пригодятся созданные при исследовании пропуска в балансе массивы с названиями столбцов непрерывных и категориальных переменных. Добавим к первым баланс и удалим из категориальных `churn`

## Статистическая проверка гипотез

В ходе сравнения признаков разных групп, мы опирались на гистограммы, которые в некоторых случаях дают однозначное численное или визуальное понимание различия признаков (например, уходит каждый пятый мужчина и каждая десятая женщина - вопросов и сомнений нет). В некоторых случаях различие не столь очевидно, либо вовсе незаметно.   
Проверим статистическими методами ряд гипотез, чтобы дополнительно обосновать различия между группами, либо указать на их отстутствие.

Проверим следующие гипотезы:   

- гипотеза различия возраста между теми клиентами, которые пользуются двумя продуктами банка, и теми, которые пользуются одним
- гипотеза равенства распределения среднего рейтинга в группах 2 и 3
- гипотеза равенства долей активных клиентов во всех группах

### Гипотеза различия возраста клиентов группы 2 и группы 1
Возраст хорошо характеризуется средним значением. Наши клиенты старше 18 лет, сверху возраст не ограничен, но фантастических выбросов ожидать не стоит всилу естественных причин.   
Для сравнения средних двух выборок используем критерий Стьюдента и стандартный уровень значимости .05

Н0: Средний возраст клиентов 1ой и 2ой группы одинаковый
Н1: Средний возраст клиентов 1ой и 2ой группы различается

```python
alpha = .05
```

```python
results = st.ttest_ind(group1['age'], group2['age'])
print('p-значение:', results.pvalue)
if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу") 
```

Несмотря на визуально схожие распределения, статистически вероятность получить две выборки нашего датасета при условии, что верна нулевая гипотеза и средний возраст двух групп одинаковый - всего 3.4%, что меньше общепринятого уровня статистической значимости.

```python
print('Средний возраст группы 1:', round(group1['age'].mean(),1))
print('Средний возраст группы 2:', round(group2['age'].mean(),1))
```

И хотя средний возраст отличается всего чуть больше чем на пол года, это различие статистически значимо.

### Гипотеза равенства распределения кредитного рейтинга в группе 2 и 3 
Если в первой группе очевиден пик кредитного рейтинга около значения в 900 пунктов, то во 2ой и 3ей группе распределения похожи.
Средний рейтинг нам не очень интересен - проверим распределение признака. По виду гистограммы они похоже, так ли это с точки зрения статистики, поможет понять критерий Манна-Уитни.   

H0: Распределение кредитного рейтинга в группе 2 и 3 одинаковое   
Н1: Распределение кредитного рейтинга в группах 2 и 3 различается

```python
p_value = st.mannwhitneyu(group1['score'], group2['score'])[1]
print('p-значение: ', p_value)
if p_value < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
```

Статистически распределения кредитного рейтинга в группах отличаются.

### Гипотеза равенства долей активных клиентов во всех группах
На глаз мы предположили, что доля активных клиентов одинакова во всех группах. Проверим с помощью критерия равенства пропорций, так ли это с точки зрения статистики. 

Н0: Между долями активных клиентов в трех группах нет значимой разницы
Н1: Разница долей активных клиентов между группами есть

```python
#Количество активных клиентов в группе 1
a1 = group1['last_activity'].sum()
#Всего клиентов в группе 1
t1 = group1['last_activity'].count()
#Аналогично для двух других групп:
a2 = group2['last_activity'].sum()
t2 = group2['last_activity'].count()
a3 = group3['last_activity'].sum()
t3 = group3['last_activity'].count()
```

Всего проведем три попарных сравнения 1-2, 1-3 и 2-3 группы. Чтобы утверждать, что различия между группами нет, сохраняя примерно тот же уровень значимости, что и в предыдущих тестах, уровень значимости уменьшим в три раза (поправка Бонферрони):

```python
alpha = alpha / 3
alpha
```

```python
#Сравнение групп 1 и 2
stat, p_value = proportions_ztest(np.array([a1,a2]), np.array([t1,t2]))
print('p-значение: ', p_value)
if p_value < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
```

```python
#Сравнение групп 1 и 3
stat, p_value = proportions_ztest(np.array([a1,a3]), np.array([t1,t3]))
print('p-значение: ', p_value)
if p_value < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
```

```python
#Сравнение групп 3 и 2
stat, p_value = proportions_ztest(np.array([a3,a2]), np.array([t3,t2]))
print('p-значение: ', p_value)
if p_value < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
```

Наше предположение подтвердилось. Значение p-value в каждом сравнении значительно выше 0.05, так что даже самая суровая поправка на попарное сравнение не пригодилась.
